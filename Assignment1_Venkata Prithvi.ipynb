```python
#14 WAPP reverse a string using while loop
input_string = input("Enter a String:") # take input from user
rev_string = ""
length = len(input_string)# we find length of the string
index= length -1
while index>=0:
    rev_string+= input_string[index] #add character at current index
    index-=1 #move one step back
print("Reversed string:" ,rev_string)

    

```
