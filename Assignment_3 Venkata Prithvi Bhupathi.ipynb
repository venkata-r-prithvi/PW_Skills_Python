```python
#6.concept of generators in Python and how they are defined.
'''
Generators : is defined like a "regular function " but uses "yeild" to produce values.
each time yeild is encountered the function pauses ,returns the current value and retains its state 
for next call.
generator object supports the iterator protocol so that it can be used in loops with next().

'''
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count+= 1
#using generator
for num in count_up_to(5):
    print(num)
```

```python
#7. advantages of using generators over regular functions?
'''
Advantages of Generators over regular functions are as follows:

Memory Efficieny: Generators produce values one at a time and "do not store all values at once".
Ex useful for processing large data sets. "Regular functions" return all values at once
Lazy evaluation : Generators compute values only when needed ,do not perform calculations upfront
Regular function execute all the code. which is inefficient for large data sets.
Infinte Sequences : Generators can produce infinite sequences. such as endless stream of numbers or
data.This is not possible with regular function. They must return value.
'''
def square_num(nums): #normal function
    result = []
    for num in nums:
        result.append(num*num)
    return result
def square_num_gen(nums): #using yeild
        for num in nums:
            yield num * num
for square in square_num_gen([1,2,3,4,5]):
    print(square)
```

```python
#4.WAP function that checks if a given number is prime or not from 1 to 200.
def prime_num(n):
    # Check if n is in the required range
    if not (1 <= n <= 200):
        print("Please enter the number between 1 and 200")
        return None

    # Check special cases
    if n == 1:
        print("1 is NOT a prime number")
        return False

    # Check divisibility
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
prime_num(789) 
```

```python
#5.WAP iterator class in Python that generates the Fibonacci sequence up to a specified number of
#terms.

class Fib_Iter:
    def __init__(self, n): # n is specified number of terms
        self.n = n              # number of terms
        self.a, self.b = 0, 1   # Initialing a &  b 
        self.count = 0

    def __iter__(self):  # Recursive function 
        return self

    def __next__(self):
        if self.count == self.n: #condition when to stop i.e to reach the  output value.
            raise StopIteration
        val = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return val


# first 10 Fibonacci numbers
for num in Fib_Iter(10):
    print(num)
```

```python
#6.WAP generator function in Python that yields the powers of 2 up to a given exponent.

def power_of_2(n): #n is the given exponent
    for i in range(n+1): # we use range of n+1 because we cant iterate integer directly
        yield 2**i #generator function uses yeild instead of return.
                   # i gives current iterator value if we give n it will give final value
```

```python
for num in power_of_2(8):
    print (num)
```

```python
#7. WAP generator function that reads a file line by line and yields each line as a string.
def read_file_lines(filename):
    """Generator that reads a file line by line and yields each line as a string."""
    with open(filename, 'r') as file: # opens file in read mode
        for line in file:
            yield line.strip()  # remove newline character and yield the line

file_path = r"C:\Users\prith\Desktop\example.txt"  # we use raw string to avoid escape errors

for line in read_file_lines(file_path):
    print(line)
```

```python
#9.WAPP to that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

def celsius_to_fahrenheit(temp_list):
    """Convert a list of temperatures from Celsius to Fahrenheit using map()."""
    return list(map(lambda c: (c * 9/5) + 32, temp_list))


temps_c = [0, 25, 37, 100]
temps_f = celsius_to_fahrenheit(temps_c)
print(temps_f)

```

```python
#10. WAPP that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(text):
    vowels = 'aeiouAEIOU'  # all vowels
    return ''.join(filter(lambda ch: ch not in vowels, text)) # keep only characters that are not vowels

# Example
input_text = "aeiou pwskills"
result = remove_vowels(input_text)
print(result)
```
